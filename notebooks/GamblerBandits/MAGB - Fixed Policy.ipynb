## MAGB $j$-Fixed Policy

Expected Frequencies:

$$
\forall t \in \mathbb{N}, \forall i \in \mathcal{I} : N_{i,t} = 
    \begin{cases}
        t & \text{ if } i = j \\
        0 & \text{ otherwise }
    \end{cases}
$$

Expected Precision:

$$
\forall t \in \mathbb{N} : \psi_t = 
    \begin{cases}
        1 & \text{ if } i = j \\
        0 & \text{ otherwise }
    \end{cases}
$$

Success Probability

$$
\forall t \in \mathbb{N} : \mathbb{P}(X_t = 1) = p_j
$$

Expected Reward

$$
\forall t \in \mathbb{N} : \mathbb{E} \left[ R_t \right] = \mu_j
$$

Probability of ruin

$$
\lim_{h \to \infty} \mathbb{P}(\tau \leq h) 
= \begin{cases}
\left( \frac{1-p_j}{p_j} \right)^b & \text{ if } p_j > 0.5 \\
1 & \text{ if } p_j \leq 0.5
\end{cases}
$$

The probability of being ruined exactly after t rounds

$$
\mathbb{P}(\tau = t) 
= \begin{cases}
    0 & \text{ if } b>t \text{ or odd}(t+b)\\
    \max \left[ 0 \ , \ \binom{t-1}{\frac{t-b}{2}} - \binom{t-1}{\frac{t-b}{2}-1} \right] \cdot p_j^{(\frac{t-b}{2})} \cdot (1-p_j)^{(\frac{t+b}{2})} & \text{ otherwise }  
  \end{cases}$$



## MAGB Uniform Random Policy

Expected Frequencies:

$$
\forall t \in \mathbb{N}, \forall i \in \mathcal{I} : \mathbb{E}[N_{i,t}] = \frac{1}{k}
$$

Expected Precision:

$$
\forall t \in \mathbb{N} : \mathbb{E}[\psi_t] = \frac{1}{k}
$$

Success Probability

$$
\forall t \in \mathbb{N} : \mathbb{P}(X_t = 1) = \bar{p} = \frac{1}{k} \sum_{i=1}^k p_i
$$

Expected Reward

$$
\forall t \in \mathbb{N} : \mathbb{E} \left[ R_t \right] = \frac{1}{k} \sum_{i=1}^k \mu_i
$$

Probability of ruin

$$
\lim_{h \to \infty} \mathbb{P}(\tau \leq h) 
= \begin{cases}
\left( \frac{1-\bar{p}}{\bar{p}} \right)^b & \text{ if } \bar{p} > 0.5 \\
1 & \text{ if } \bar{p} \leq 0.5
\end{cases}
$$

The probability of being ruined exactly after t rounds

$$
\mathbb{P}(\tau = t) 
= \begin{cases}
    0 & \text{ if } b>t \text{ or odd}(t+b)\\
    \max \left[ 0 \ , \ \binom{t-1}{\frac{t-b}{2}} - \binom{t-1}{\frac{t-b}{2}-1} \right] \cdot (\bar{p})^{(\frac{t-b}{2})} \cdot (1-\bar{p})^{(\frac{t+b}{2})} & \text{ otherwise }  
  \end{cases}$$


## $w$-Explore-Then-Commit (ETC)

Plays each arm exactly $w$ times, sequentially, then, after $wk$ rounds, fix the choice to the best action in that round.
Note that it is not greedy. After $wk$ rounds, the estimated mean is no longer calculated, and the action does not change anymore.

Expected Frequencies:

$$
\forall t > wk : N_{i,t} = 
    \begin{cases}
        w & \text{ if } i \neq i^* \text{ at } t=wk \\
        w + (t-wk) & \text{ if } i = i^* \text{ at } t=wk
    \end{cases}
$$

Expected Precision:

$$
\forall t \in \mathbb{N} : \psi = 
    \begin{cases}
        1 & \text{ if } i = j \\
        0 & \text{ otherwise }
    \end{cases}
$$

Success Probability

$$
\forall t \in \mathbb{N} : \mathbb{P}(X_t = 1) = p_j
$$

Expected Reward

$$
\forall t \in \mathbb{N} : \mathbb{E} \left[ R_t \right] = \mu_j
$$

Probability of ruin

$$
\lim_{h \to \infty} \mathbb{P}(\tau \leq h) 
= \begin{cases}
\left( \frac{1-p_j}{p_j} \right)^b & \text{ if } p_j > 0.5 \\
1 & \text{ if } p_j \leq 0.5
\end{cases}
$$

The probability of being ruined exactly after t rounds

$$
\mathbb{P}(\tau = t) 
= \begin{cases}
    0 & \text{ if } b>t \text{ or odd}(t+b)\\
    \max \left[ 0 \ , \ \binom{t-1}{\frac{t-b}{2}} - \binom{t-1}{\frac{t-b}{2}-1} \right] \cdot p_j^{(\frac{t-b}{2})} \cdot (1-p_j)^{(\frac{t+b}{2})} & \text{ otherwise }  
  \end{cases}$$



## MAGB Greedy Empirical Means

Supposing each arm is pulled once,
we have $2^k$ possible combinations of observations after $k$ rounds.

Let $c$ be the index of a combination.

The probability of observing a given combination is:

$$
\mathbb{P}(H_k = c) = 
\prod_{i = 1}^{k}
\begin{cases}
p_i & \text{ if odd} (c//(2^{(i-1)}))\\
1-p_i & \text{ otherwise}
\end{cases}
$$

The best $p$ (optimistic) for a combination $c$ is:

$$
\max(p_i \mid c) = \max_{1 \leq i \leq k}
\begin{cases}
p_i & \text{ if odd} (c//(2^{(i-1)}))\\
0 & \text{ otherwise}
\end{cases}
$$

The worst $p$ (pessimistic) for a combination $c$ is:

$$
\min(p_i \mid c) = 
\min_{1 \leq i \leq k}
\begin{cases}
p_i & \text{ if odd} (c//(2^{(i-1)}))\\
1 & \text{ otherwise}
\end{cases}
$$

Optimistic Bound :
$$
\sum_{c = 0}^{2^k}
\mathbb{P}(H_k = c)
\max(p_i \mid c)
$$

Pessimist Bound :
$$
\sum_{c = 0}^{2^k}
\mathbb{P}(H_k = c)
\min(p_i \mid c)
$$


When Greedy Empirical Means is combined with Bernoulli rewards, the arms that returns a failure in the first observation will never be tried again, except in the particular case when all arms return a failure. It can be repeated recursivelly until the roun when at least one arm returns a success.

$$ \mathbb{P}(S_{i,k} = 0) = (1-p_i) $$

$$ \mathbb{P}(S_{j,k} = 0, \forall j \neq i) = \prod_{j \neq i} (1-p_j) $$

$$ \mathbb{P}(\exists j \neq i : S_{j,k} = 1) = 1 - \prod_{j \neq i} (1-p_j) $$

Not that:  ...

$$ \mathbb{P}(S_{i,wk} = 0) = \sum_{t=1}^w \left( \prod_{j = 1}^k (1 - p_j) \right)^{t-1} (1-p_i) (1-\prod_{j \neq i} (1 - p_j))$$

```python
################################################################################
from math import comb
from decimal import Decimal, getcontext
from fractions import Fraction

getcontext().prec = 28

def zcomb(n:int, k:int):
    #k and n must be naturals
    if (n >= k) and (k >= 0):
        return comb(n, k)
    else:
        return 0


################

def big_int_to_float(v, power=0):
    try:
        return float(v // 2**power), power
    except:
        return big_int_to_float(v, power+1)

def exact_ruin_dist(t, b, p=0.5):
    if (abs(b)>t) or ((t+b)%2):
        return 0
    else:
        #centered catalan triangle
        k = (t+b)//2
        n = k-1 #(t+b-2)//2
        m = (t-b)//2
        c = max(0, zcomb(n+m, m) - zcomb(n+m, m-1))
        try:
            return c * p**(t-k) * (1-p)**k
        except:
            return Decimal(c) * Decimal(p)**(t-k) * Decimal((1-p))**k

        #from scipy.stats import binom
        #rv = binom(n+m, p)
        #rv.pmf(m)
        #rv.pmf(m-1)
        
################

def ruin_dist(h, b, p=0.5):
    s = 0.0
    for t in range(h):
        s += exact_ruin_dist(t, b, p)
    return s    

###############
#greedy analysis

pr_arr = [1.0] * 2**k
max_p_arr = [-2.0] * 2**k
min_p_arr = [+2.0] * 2**k
for c in range(0, 2**k):
    for i in range(1, k+1):
        p = p_arr[i-1]
        if (c//(2**(i-1)))%2 == 1:
            pr_arr[c] = pr_arr[c] * p
            max_p_arr[c] = max(max_p_arr[c], p)
            min_p_arr[c] = min(min_p_arr[c], p)
        else:
            pr_arr[c] = pr_arr[c] * (1-p)
    
pr = pr_arr[0] / ((2**k)-1)
for c in range(1, 2**k):
    pr_arr[c] += pr
pr_arr[0] = 0.0

opt_p = 0.0
pes_p = 0.0
for c in range(0, 2**k):
    opt_p += pr_arr[c] * max_p_arr[c]
    pes_p += pr_arr[c] * min_p_arr[c]

###############

p_arr_ext = p_arr.copy()

#random, optimistic on greedy, pessimistic on greedy
#p_arr_ext = p_arr.copy() + [np.mean(p_arr), opt_p, pes_p]
#print(p_arr_ext)

def prob_ruin_inf(p, b_0):
    return ((1-p)/p)**b_0  if  p > 0.5  else  1.0

prob_ruin_at = np.zeros((len(p_arr_ext),h))
for i in range(len(p_arr_ext)):
    p = p_arr_ext[i]
    for t in range(h):
        prob_ruin_at[i,t] = exact_ruin_dist(t, b_0, p)
        
prob_ruin_until = np.cumsum(prob_ruin_at, axis=1)

expected_reward_at = np.zeros((len(p_arr_ext),h))
for i in range(len(p_arr_ext)):
    p = p_arr_ext[i]
    for t in range(h):
        expected_reward_at[i,t] = (1.0 - prob_ruin_until[i,t]) * (2*p-1)

expected_cum_reward = np.cumsum(expected_reward_at, axis=1)
        

###############

fig, ax = plt.subplots(1)
sim.plot(np.array([[1-prob_ruin_inf(p, b_0)]*h for p in p_arr_ext]), title='Expected Survival Rate', ax=ax, line_properties=[{'ls':':', 'label':''} for _ in p_arr_ext], show=False, showlast='axis')
sim.plot(1-prob_ruin_until, ax=ax, showlast='legend')
sim.plot(expected_reward_at, title='Expected Instant Reward (considering ruin)')
sim.plot(expected_cum_reward, title='Expected Cumulated Reward (considering ruin)')
sim.plot(expected_cum_reward + b_0, title='Expected Budget (considering ruin)')
#sim.plot(expected_cum_reward + b_0, title='Expected Precision (considering ruin)')
```
