# Gambler Automata

Extending the classic Gambler's Ruin problem to multi-state


```python
import networkx as nx
import matplotlib.pyplot as plt

graph = nx.DiGraph()

nodes = ["$0$", "$1$", "...", "$b_0$", "....", "$g-1$", "$g$"]
pos = {node:(i, 1) for i, node in enumerate(nodes)}
edges = [("$0$", "$1$"), ("$1$", "$0$"), 
         ("$1$", "..."), ("...", "$1$"), 
         ("...", "$b_0$"), ("$b_0$", "..."), 
         ("$b_0$", "...."), ("....", "$b_0$"), 
         ("....", "$g-1$"), ("$g-1$", "...."), 
         ("$g-1$", "$g$"), ("$g$", "$g-1$")]

graph.add_nodes_from(nodes) 
graph.add_edges_from(edges) 

fig, ax = plt.subplots(figsize=(12, 3))

nx.draw_networkx(graph, pos=pos, connectionstyle='arc3, rad = 0.3', node_size=1500, node_color='orange')

plt.show()
                     
```

## First Passage

### Random Walk in Infinite Linear Markov Chain

if the chain is on $\mathbb{Z}$, from $-\infty$ to $\infty$, with step $1$,

startig from $0$ and targeting $g$ :

considering $g > 0$ :

$
\mathbb{P}(\exists t \mid s_t = g) \ \ = \
\begin{cases}
1 & \text{if } \ p \geq q \\
\left(\frac{p}{q}\right)^g & \text{if } \ p < q
\end{cases}
$

$
\mathbb{E}(t \mid s_t = g) \ \ = \
\begin{cases}
\infty & \text{if } \ p \leq q \\
\frac{g}{p-q} & \text{if } \ p < q
\end{cases}
$
