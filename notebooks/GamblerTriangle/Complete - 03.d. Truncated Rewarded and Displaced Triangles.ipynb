# Truncating Displaced Rewarded Triangles

## $h$-Truncated $(d,b)$-Displaced Centered Triangle

$\mathcal{T}\!\mathcal{runc}_h \big[ Z_{d,b} \big](n, s) = 
\left[ \binom{n-d}{\frac{n-d+s-b}{2}} - \binom{n-d}{\frac{n-d+s+b}{2} - h} \right] 
\cdot [ (n-d+s-b+1) \text{ mod } 2 ] $


## $h$-Truncated $(r,c)$-Rewarded Triangle

$\begin{align}
\mathcal{T}\!\mathcal{runc}_h \big[ Z^{r,c}\big](n, s) 
& \ = \ Z^{r,c}(n,s) - Z_{h - \frac{ch}{r}}^{r,c}(n,s)\\
& \ = \ \begin{cases}
\binom{n}{\frac{s-cn}{r-c}} - \binom{n}{\frac{s+h-\frac{ch}{r}-cn}{r-c}} & \text{if } \ \frac{s-cn}{r-c}, \frac{h}{r} \in \mathbb{N} \\
0 & \text{otherwise}
\end{cases} 
\end{align}$

## $h$-Truncated $b$-Budgeted $(r,c)$-Rewarded Triangle

$\begin{align}
\mathcal{T}\!\mathcal{runc}_h \big[ Z_b^{r,c}\big](n, s) 
& \ = \ 
Z_b^{r,c}(n,s) - Z_{b + h - \frac{ch}{r}}^{r,c}(n,s)\\
& \ = \ 
\begin{cases}
\binom{n}{\frac{s-b-cn}{r-c}} - \binom{n}{\frac{s-b+h-\frac{ch}{r}-cn}{r-c}} & \text{if } \ \frac{s-b-cn}{r-c}, \frac{h}{r} \in \mathbb{N} \\
0 & \text{otherwise}
\end{cases}
\end{align}$

## $h$-Truncated $(d,b)$-Translated $(r,c)$-Rewarded Triangle

$\begin{align}
\mathcal{T}\!\mathcal{runc}_h \big[ Z_{d,b}^{r,c}\big](n, s)
& \ = \ 
Z^{r,c}_{d,b}(n,s) - Z_{d, b + h - \frac{ch}{r}}^{r,c}(n,s)\\
& \ = \ 
\begin{cases}
\binom{n-d}{\frac{s-b-cn+dc}{r-c}} - \binom{n-d}{\frac{s-b+h-\frac{ch}{r}-cn+dc}{r-c}} & \text{if } \ \frac{s-b-cn+dc}{r-c}, \frac{h}{r} \in \mathbb{N} \\
0 & \text{otherwise}
\end{cases}
\end{align}$


```python
from pyrl.gr import *
from pyrl.utils import printdf

################

max_s = 20
s_arr = range(-max_s,max_s+1)
n_arr = range(-3,max_s+1)

printdf([[trunc_center_triangle(n, s, -1)     for s in s_arr] for n in n_arr], label_rows=n_arr, label_cols=s_arr, label_axis_cols="s", label_axis_rows="n", title=f'0-Centered $-1$-Truncated Pascal Triangle:', transpose=True, reverse=True)
printdf([[trunc_center_triangle(n, s, -3)    for s in s_arr] for n in n_arr], label_rows=n_arr, label_cols=s_arr, label_axis_cols="s", label_axis_rows="n", title=f'0-Centered $-3$-Truncated Pascal Triangle:', transpose=True, reverse=True)
printdf([[trunc_center_triangle(n, s, +7)    for s in s_arr] for n in n_arr], label_rows=n_arr, label_cols=s_arr, label_axis_cols="s", label_axis_rows="n", title=f'0-Centered $+7$-Truncated Pascal Triangle:', transpose=True, reverse=True)

printdf([[trunc_budget_triangle(n, s, 3, 5)    for s in s_arr] for n in n_arr], label_rows=n_arr, label_cols=s_arr, label_axis_cols="s", label_axis_rows="n", title='+5-Centered +3-Truncated Pascal Triangle:', transpose=True, reverse=True)
printdf([[trunc_budget_triangle(n, s, 5, 3)    for s in s_arr] for n in n_arr], label_rows=n_arr, label_cols=s_arr, label_axis_cols="s", label_axis_rows="n", title='+3-Centered +5-Truncated Pascal Triangle:', transpose=True, reverse=True)
printdf([[trunc_budget_triangle(n, s, -3, -5)  for s in s_arr] for n in n_arr], label_rows=n_arr, label_cols=s_arr, label_axis_cols="s", label_axis_rows="n", title='-5-Centered -3-Truncated Pascal Triangle:', transpose=True, reverse=True)
printdf([[trunc_budget_triangle(n, s, -5, -3)  for s in s_arr] for n in n_arr], label_rows=n_arr, label_cols=s_arr, label_axis_cols="s", label_axis_rows="n", title='-3-Centered -5-Truncated Pascal Triangle:', transpose=True, reverse=True)
```
